```python
class FourCal:
  def __init__(self, first, second):
    self.first = first
    self.second = second
  
  def setdata(self, first, second):
    self.first = first
    self.second = second
  
  def add(self):
    result = self.first + self.second
    return result

  def sub(self):
    result = self.first - self.second
    return result
  
  def mul(self):
    result = self.first * self.second
    return result

  def div(self):
    result = self.first / self.second
    return result
```

* self에는 setdata메서드를 호출한 객체가 자동으로 전달 
* init은 생성자

```python
class MoreFourCal(FourCal):
  def pow(self):
    result = self.first ** self.second
    return result
```

* 상속을 할 때는 위와 같이 클래스명 옆에 상속할 클래스 이름을 input
* overriding 역시 가능
